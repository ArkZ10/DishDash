```python
ratings_with_food = ratings.merge(food_data, on='food_id')
```

```python
ratings_with_food.shape
```

```python
number_rating = ratings_with_food.groupby('food_name')['rating'].count().reset_index()
```

```python
number_rating.head(10)
```

```python
number_rating.rename(columns={'rating':'sum_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_food.merge(number_rating, on='food_name')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['sum_of_rating'] >= 250]
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','food_name'],inplace=True)
final_rating.shape
```

```python
food_pivot = final_rating.pivot_table(columns='user_id', index='food_name', values= 'rating')
food_pivot
```

```python
food_pivot.shape
```

```python
food_pivot.fillna(0, inplace=True)
food_pivot
```

### TRAINNNNNN

```python
from scipy.sparse import csr_matrix
```

```python
food_sparse = csr_matrix(food_pivot)
```

```python
type(food_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(food_sparse)
```

```python
distance, recommendation = model.kneighbors(food_pivot.iloc[230,:].values.reshape(1,-1), n_neighbors=6 )
```
