```python
data = data[['food_id', 'Nama Bahan Makanan', 'Fast_Food_New', 'Tipe_New',
             'Beras', 'Biji-bijian', 'Buah', 'Daging Ayam', 'Daging Babi',
             'Daging Kambing', 'Daging Sapi', 'Ikan', 'Kedelai', 'Sayur',
             'Susu', 'Telur Ayam', 'Tepung', 'Umbi-umbian']]

data.columns
data.shape
```

```python
new_data = np.array(data.iloc[:,3:])
dot_prod_new_data = new_data .dot(np.transpose(new_data))
```

```python
dot_prod_new_data
```

```python
def find_similar_food(ID):
    '''
    INPUT
    Food_ID
    OUTPUT
    similar food
    '''
    food_idx = np.where(data['food_id'] == ID)[0][0]
    similar_idxs = np.where(dot_prod_new_data[food_idx] == np.max(dot_prod_new_data[food_idx]))[0]
    # similar_idxs = np.where(dot_prod_new_data[food_idx] >=3 )[0]


    similar_food = np.array(data.iloc[similar_idxs, ]['Nama Bahan Makanan'])

    return similar_food
```

```python
food_id = 1
food = data.at[food_id,'Nama Bahan Makanan']
similar_food = find_similar_food(food_id)
simplified = []
for i in similar_food[0:5]:
  simplified.append(i)


print(f'Makanan Pilihan: {food}\nMakanan Rekomendasi = {simplified}')
```
